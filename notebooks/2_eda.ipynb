```python
# 1 City-wise Average AQI

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../cleaned_data/cleaned_city_day.csv")
df["Date"] = pd.to_datetime(df["Date"])

city_avg = df.groupby("City")["AQI"].mean().sort_values(ascending=False)

plt.figure(figsize=(12,5))
city_avg.head(10).plot(kind="bar")
plt.title("Top 10 Most Polluted Cities (Average AQI)")
plt.ylabel("AQI")
plt.show()
```

```python
# 2 Monthly AQI Seasonality

df["Month"] = df["Date"].dt.month

monthly_avg = df.groupby("Month")["AQI"].mean()

plt.figure(figsize=(10,5))
plt.plot(monthly_avg.index, monthly_avg.values, marker="o")
plt.title("Monthly AQI Trend (Seasonality)")
plt.xlabel("Month")
plt.ylabel("Average AQI")
plt.show()
```

```python
# 3 Correlation Heatmap

import seaborn as sns

plt.figure(figsize=(10,6))
sns.heatmap(df[['PM2.5','PM10','NO2','SO2','CO','O3','AQI']].corr(),
            annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Pollutants")
plt.show()
```

```python
# 4 AQI Bucket Distribution

# if this cell is run alone, ensure df is defined
if 'df' not in globals():
	import pandas as pd
	import matplotlib.pyplot as plt
	df = pd.read_csv("../cleaned_data/cleaned_city_day.csv")
	df["Date"] = pd.to_datetime(df["Date"])
	df["Month"] = df["Date"].dt.month

bucket_count = df["AQI_Bucket"].value_counts()

plt.figure(figsize=(10,5))
bucket_count.plot(kind="bar")
plt.title("AQI Category Distribution")
plt.ylabel("Count")
plt.show()
```
