```python
import pandas as pd
import numpy as np

df = pd.read_csv("../datasets/city_day.csv")

df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values(["City", "Date"])
df = df.drop_duplicates()

pollutants = ['PM2.5','PM10','NO','NO2','NOx','NH3','CO','SO2','O3',
              'Benzene','Toluene','Xylene','AQI']

df[pollutants] = df.groupby("City")[pollutants].transform(lambda x: x.interpolate())

df[pollutants] = df.groupby("City")[pollutants].transform(lambda x: x.fillna(x.mean()))

df.to_csv("../cleaned_data/cleaned_city_day.csv", index=False)

print("Cleaned dataset saved successfully!")
```
