```python
import pandas as pd

arima = pd.read_csv("../forecasts/arima_forecast_all_cities.csv")
prophet = pd.read_csv("../forecasts/prophet_forecast_all_cities.csv")

arima["Model"] = "ARIMA"
prophet["Model"] = "PROPHET"

final_forecast = pd.concat([arima, prophet])
final_forecast.to_csv("../forecasts/final_forecast.csv", index=False)

print("final_forecast.csv created successfully!")
```
